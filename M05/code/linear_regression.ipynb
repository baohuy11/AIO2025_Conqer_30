```python
import numpy as np
import matplotlib.pyplot as plt

# Data
X = np.array([[1, 50], [1, 80], [1, 100], [1, 120]])
y = np.array([[1500], [2400], [3000], [3600]])

# Apply Normal Equation
theta = np.linalg.inv(X.T @ X) @ X.T @ y

# Result: theta[0] = intercept, theta[1] = slope (gradient)
print("Optimal parameters:", theta) 
```

```python
intercept = theta[0][0]
slope = theta[1][0]
```

```python
X_area = X[:, 1]

x_min = X_area.min() - 10
x_max = X_area.max() + 10
X_line = np.linspace(x_min, x_max, 100)
```

```python
y_pred = intercept + slope * X_line
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(X_area, y, color='#4c51bf', s=100, label='(Actual Data)')

plt.plot(X_line, y_pred, color='#e53e3e', linewidth=3, 
         label=f'ng hi quy: Gi = {slope:.2f} * Din tch + {intercept:.2e}')

plt.title('Linear Regression')
plt.xlabel('Din tch ($m^2$)')
plt.ylabel('Gi (triu ng)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)


plt.tight_layout()
plt.show()
```
